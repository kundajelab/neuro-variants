```python
import pandas as pd
```

```python
input_variants_file = '/oak/stanford/groups/akundaje/projects/neuro-variants/variant_lists/excitatory_neurons_dev.dataset.txt'
rare_variants_file = '/oak/stanford/groups/akundaje/projects/neuro-variants/variant_lists/neuro.rare.variants.1kg.lt_0.001.tsv'
common_variants_file = '/oak/stanford/groups/akundaje/projects/neuro-variants/variant_lists/neuro.common.variants.1kg.gt_0.05.tsv'
output_variants_file = '/oak/stanford/groups/akundaje/projects/neuro-variants/variant_lists/shap_variants.rare.model_inputs.hg38.tsv'
```

```python
input_variants = pd.read_table(input_variants_file)

input_variants
```

```python
rare_variants = pd.read_table(rare_variants_file, names=['chr_hg38', 'pos_hg38', 'ref_allele', 'alt_allele', 'snp_id'])

rare_variants
```

```python
output_variants = rare_variants.loc[rare_variants['snp_id'].isin(input_variants['snp_id'])].copy()

output_variants
```

```python
input_variants.loc[~input_variants['snp_id'].isin(output_variants['snp_id'])]
```

```python
common_variants = pd.read_table(common_variants_file, names=['chr_hg38', 'pos_hg38', 'ref_allele', 'alt_allele', 'snp_id'])

common_variants
```

```python
overlap_variants = common_variants.loc[common_variants['snp_id'].isin(input_variants['snp_id'])].copy()

overlap_variants
```

```python
output_variants.to_csv(output_variants_file, sep='\t', header=False, index=False)
```
